```python
import json as js
student_grades = {
    "Rahul": {"Math": 85, "Science": 90, "English": 78},
    "Priya": {"Math": 92, "Science": 88, "English": 95},
    "Amit": {"Math": 76, "Science": 82, "English": 80}
}
def save_grades_to_json(filename,grades_dict):
    with open (filename,'w') as f:
        js.dump(grades_dict,f)
    

save_grades_to_json("/Users/sagar.s.sarambalkar/github/assignments/AssignmentsMasai/studentrecordsjson/grades.json", student_grades) 

def load_grades_from_json(filename):
    with open(filename,'r') as f:
        return js.load(f)
    
data=load_grades_from_json("/Users/sagar.s.sarambalkar/github/assignments/AssignmentsMasai/studentrecordsjson/grades.json")
#print(f"{data.items()}")
for student,subjects in data.items():
    avg=sum(subjects.values()) / len(subjects)
    print(f"student name :{student}  subjects avg : {avg}")

   
```
