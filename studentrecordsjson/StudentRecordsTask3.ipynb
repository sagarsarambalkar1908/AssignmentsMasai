```python
student_record1 = "Neha,Present,2024-01-15"
student_record2="Vikram,Absent,2024-01-15"
def add_attendance_record(filename, student_record):
    with open (filename,'a') as f:
        f.write(student_record+"\n")

add_attendance_record("/Users/sagar.s.sarambalkar/github/assignments/AssignmentsMasai/studentrecordsjson/attendance.txt", student_record1)
add_attendance_record("/Users/sagar.s.sarambalkar/github/assignments/AssignmentsMasai/studentrecordsjson/attendance.txt", student_record2)

def count_present_students(filename):
    count=0
    with open (filename,'r') as f:
        records=f.readlines()
        print(type(records))
        for record in records:
            data=record.split(',')
            if data[1] == "Present":
                count=count+1
    return count


total_students=count_present_students("/Users/sagar.s.sarambalkar/github/assignments/AssignmentsMasai/studentrecordsjson/attendance.txt")
print(f"Total Present Student : {total_students}")


        
```
